```python
df = pd.DataFrame(jogadores)
df.columns = ["Nome do Jogador", "Posição", "Idade", "Minutos Jogados", "Partidas Jogadas", "Gols Marcados", "Assitências", "Desarmes", "Chutes Defendidos", "Porcentagem de Passes"]
df
```

```python
model = "models/embedding-001"
```

```python
def embed_fn(row):
  texto = f"""Jogador: {row['Nome do Jogador']},
  Posição: {row['Posição']},
  Idade: {row['Idade']},
  Minutos Jogados: {row['Minutos Jogados']},
  Partidas Jogadas: {row['Partidas Jogadas']},
  Gols Marcados: {row['Gols Marcados']},
  Assistências: {row['Assitências']},
  Desarmes: {row.get('Desarmes', 0)},
  Chutes Defendidos: {row.get('Chutes Defendidos', 0)}"""

  return genai.embed_content(model=model, content=texto, task_type="RETRIEVAL_DOCUMENT")["embedding"]

df["Embeddings"] = df.apply(embed_fn, axis=1)
df
```

```python
def gerar_e_buscar_consulta(consulta, base, model):
  """
  Gera embeddings para a consulta, busca em uma base de dados e retorna resultados relevantes.

    Args:
        consulta: A consulta do usuário.
        base: A base de dados onde a busca será realizada.
        model: O modelo de embeddings a ser utilizado.

    Returns:
        Uma lista de resultados relevantes da base de dados.
  """
  embedding_da_consulta = genai.embed_content(model=model,
                                 content=consulta,
                                 task_type="RETRIEVAL_QUERY")["embedding"]

  produtos_escalares = np.dot(np.stack(df["Embeddings"]), embedding_da_consulta)

  indice = np.argmax(produtos_escalares)
  return df.iloc[indice][["Nome do Jogador", "Posição", "Idade", "Minutos Jogados", "Partidas Jogadas", "Gols Marcados", "Assitências", "Desarmes", "Porcentagem de Passes"]]
```

```python
Questionamento = "Quem tem mais passes?"
consulta = Questionamento
trecho = gerar_e_buscar_consulta(consulta, df, model)
print(trecho)
```
